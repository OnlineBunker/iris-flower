```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/IRIS.csv')
iris
```

```python
iris.groupby('species').mean()
```

```python
sns.scatterplot(x='sepal_length', y='sepal_width', data=iris, hue='species')
```

```python
sns.lineplot(x='sepal_length', y='sepal_width', data=iris, hue='species')
```

```python
sns.lineplot(data=iris.drop(['species'], axis=1))
plt.show()
```

```python
iris.plot.hist(subplots=True, bins=20, figsize=(10, 8), layout=(2,2))
```

```python
sns.heatmap(iris.drop(['species'], axis=1).corr(), annot=True)
```

```python
species_name = iris['species'].unique()

for specie in species_name:
  species = iris[iris['species'] == specie].drop('species', axis=1)

  sns.heatmap(species.corr(), annot=True)
  plt.show()
```

```python
g = sns.FacetGrid(iris, hue="species", height=4, col="species")
g.map(sns.kdeplot, 'sepal_length')
g.add_legend()
plt.show()
```

```python
sns.pairplot(iris, hue='species')
```
